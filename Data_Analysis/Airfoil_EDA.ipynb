Imports

```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
from sklearn.impute import SimpleImputer
from scipy import stats
import plotly.express as px
from sklearn.model_selection import train_test_split 
```

Load Dataset

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00291/airfoil_self_noise.dat"
column_names = ["Frequency", "Angle of Attack", "Chord Length", "Free-stream Velocity", "Suction Side Displacement Thickness", "Sound Pressure Level"]
df = pd.read_csv(url, sep="\t", header=None, names=column_names)

# Display the first few rows of the dataset
print(df.head())
```

Initial Inspection

```python
print("Shape")
print(df.shape)
print(df.shape)
print("Basic Info of dataset")
print(df.info())
print("Statistical Summary of data: ")
print(df.describe(include='all'))
print("Missing Values (if any)")
print(df.isna().sum())
#No Missing Values found in this Dataset
```

Correlation Matrix

```python
corr_matric = df.corr()
sns.histplot(data=corr_matric, x="Frequency")
```

Distribution Plot: 

```python
for col in df.columns:
    plt.figure(figsize = (8,8))
    sns.histplot(data=df[col].dropna())
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

Box Plot

```python
for col in df.columns:
    plt.figure(figsize = (2,2))
    sns.boxplot(data=df[col].dropna())
    plt.title(f"BoxPlot of {col}")
    plt.xlabel(col)
    plt.show()
```

Scatter Plot

```python
fig = px.scatter(df)
fig = fig.update_layout(title="Scatter Plot of Airfoil")
fig.show()
```
