Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import mnist
import tensorflow as tf
import numpy as np
from PIL import Image
```

Loading the MNIST Dataset

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train),(x_test, y_test) = mnist.load_data()
```

Initial Inspection

```python
print(f"Training set shape: {x_train.shape}, Training labels shape: {y_train.shape}")
print(f"Test set shape: {x_test.shape}, Test labels shape: {y_test.shape}")
num_classes = len(np.unique(y_train))
print(f"Number of classes: {num_classes}")
print(f"Classes: {np.unique(y_train)}")
```

Pixel Intensity Distribution

```python
plt.figure(figsize=(10, 6))
plt.hist(x_train.flatten(), bins=50, color='blue', alpha=0.7)
plt.title("Pixel Intensity Distribution in Training Set")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

Mean and Standard Deviation of pixel distribution

```python
mean_image = np.mean(x_train, axis=0)
std_image = np.std(x_train, axis=0)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(mean_image, cmap='gray')
plt.title("Mean Pixel Intensity")
plt.subplot(1, 2, 2)
plt.imshow(std_image, cmap='gray')
plt.title("Standard Deviation of Pixel Intensity")
plt.suptitle("Mean and Standard Deviation of Pixel Intensities", fontsize=16)
plt.show()
```

Heatmap of correlation between pixel intensities 

```python
corr_matrix = np.corrcoef(x_train.reshape(x_train.shape[0], -1), rowvar=False)
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap of Pixel Intensities")
plt.show()
```
