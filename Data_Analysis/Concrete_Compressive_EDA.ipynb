Imports

```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the Dataset

```python
concrete_compressive_strength = fetch_ucirepo(id = 165)
X = concrete_compressive_strength.data.features
y = concrete_compressive_strength.data.targets
df = pd.concat([X,y],axis=1)
print(df.head(0))
```

Initial Inspection

```python
print(f'Feature Columns: {X.columns}')
print(f'Target Columns: {y.columns}')
```

```python
print(f'First few rows of the features: {X.head}')
print(f'First few rows of the target: {y.head} ')
```

```python
# Basic info. of the data
print(f'Basic info of dataset: {df.info}')
```

Descriptive Analysis 

```python
print(df.describe)
```

```python
# Missing Values
print(f'Missing Values for each column: {df.isnull().sum()}')
```

Exploratory Data Analysis

```python
# Histplot for all columns
print(f'Columns: {df.columns}')
for i in df.columns.drop('Concrete compressive strength'):
    plt.figure(figsize=(4,4))
    sns.histplot(df[i],kde=True,bins=30)
    plt.title(f'Histplot for {i}')
    plt.show()
```

```python
# Boxplot
print(f'Columns {df.columns}')
for i in df.columns.drop('Concrete compressive strength'):
    plt.figure(figsize=(4,4))
    sns.boxplot(df[i])
    plt.title(f'Boxplot for {i}')
    plt.show()
```

```python
#Heatmap
print(f'Columns {df.columns}')
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cmap='coolwarm',annot=True)
plt.title("Heatmap")
plt.show()
```

```python
# Pairplot
print(f'Columns: {df.columns}')
fig = px.scatter(df)
fig.show()
```
