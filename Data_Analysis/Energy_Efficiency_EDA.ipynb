Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo
import plotly.express as px
```

Loading Dataset

```python
energy = fetch_ucirepo(id=242)
X = energy.data.features
y = energy.data.targets
df = pd.concat([X,y],axis=1)
```

Initial Inspection

```python
print(df.head)
print(df.info)
```

```python
print(df.columns)
```

Descriptive Analysis

```python
print(df.describe(include="all"))
```

```python
# Missing Value
print(df.isnull().sum())
```

Exploratory Data Analysis

```python
# Histplot
print(f'Columns {df.columns}')
for i in df.columns:
    plt.figure(figsize=(4,4))
    sns.histplot(df[i],kde=True,bins=30)
    plt.title(f'Histplot for {i}')
    plt.show()
```

```python
# Boxplot
print(df.columns)
for i in df.columns:
    plt.figure(figsize=(4,4))
    sns.boxplot(df[i])
    plt.title(f'Box plot for {i}')
    plt.show()
```

```python
#Heatmap
print(f'Columns: {df.columns}')
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cmap='coolwarm',annot=True)
plt.title("Heatmap")
plt.show()
```

```python
# Pairplot
print(f'Columns {df.columns}')
fig = px.scatter(df)
fig.show()
```
